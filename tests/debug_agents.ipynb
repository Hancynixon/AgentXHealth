```python
# Load dataset
df = pd.read_csv("../data/raw/diabetes.csv")

# Initialize agents
lab = LabAgentIntelligent()
physical = PhysicalAgentIntelligent()
demo = DemographicAgentIntelligent()

# Fit agents
lab.fit(df, df["Outcome"])
physical.fit(df, df["Outcome"])
demo.fit(df, df["Outcome"])

print("Agents fitted")

# Create extreme test case
sample = pd.DataFrame([{
    "Glucose": 198,
    "Insulin": 310,
    "BMI": 36.9,
    "BloodPressure": 96,
    "Age": 52,
    "Pregnancies": 6
}])

# Predict
lab_out = lab.predict(sample)
physical_out = physical.predict(sample)
demo_out = demo.predict(sample)

print("LAB OUTPUT:", lab_out)
print("PHYSICAL OUTPUT:", physical_out)
print("DEMOGRAPHIC OUTPUT:", demo_out)
```
